```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 21
p = 0.5

# Generate values for k (number of successes)
k_values = np.arange(0, n+1)

# Calculate the probability mass function (PMF) for each k
pmf_values = binom.pmf(k_values, n, p)

# Plotting
plt.bar(k_values, pmf_values, color='skyblue', alpha=0.7)
plt.title('Binomial Distribution (n={}, p={})'.format(n, p))
plt.xlabel('Number of CB chromosomes')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
for i in range(len(pmf_values)):
    print(f"probability have {i} chromosome left is {pmf_values[i]}")
    print('given 50 progeny')
    print(f"progeny have {i} chromosome left is {50*pmf_values[i]}\n")
    
```
